# VECM Example
- Using a VECM to predict FANG stocks

```python
import pandas as pd
import numpy as np
import pandas_datareader as pdr
from scalecast.Forecaster import Forecaster
from scalecast.MVForecaster import MVForecaster
from scalecast.util import find_optimal_lag_order, find_optimal_coint_rank
from scalecast.auxmodels import vecm
from scalecast import GridGenerator
import matplotlib.pyplot as plt
```

## Download data using a public API

```python
FANG = [
    'META',
    'AMZN',
    'NFLX',
    'GOOG',
]

fs = []
for sym in FANG:
    df = pdr.get_data_yahoo(sym)
    df = df.asfreq('B', method='ffill')
    f = Forecaster(
        y=df['Close'],
        current_dates = df.index,
        future_dates = 65,
        end = '2022-09-30',
    )
    fs.append(f)
    
mvf = MVForecaster(*fs,names=FANG)
mvf.set_test_length(65)
mvf.set_validation_metric('mae')
mvf.add_sklearn_estimator(vecm,'vecm')

mvf
```

```python
mvf.plot()
plt.show()
```

## Augmented Dickey Fuller Tests to Confirm Unit-1 Roots

```python
for stock, f in zip(FANG,fs):
    adf_result = f.adf_test(full_res=True)
    print('the stock {} is {}stationary at level'.format(
        stock,
        'not ' if adf_result[1] > 0.05 else ''
    )
    )
```

```python
for stock, f in zip(FANG,fs):
    f.diff()
    adf_result = f.adf_test(full_res=True)
    print('the stock {} is {}stationary at its first difference'.format(
        stock,
        'not ' if adf_result[1] > 0.05 else ''
    )
    )
```

## Measure IC to Find Optimal Lag Order
- this is used to run the cointegration test

```python
lag_test = find_optimal_lag_order(mvf,train_only=True)
pd.DataFrame(
    {
        'aic':lag_test.aic,
        'bic':lag_test.bic,
        'hqic':lag_test.hqic,
        'fpe':lag_test.fpe,
    },
    index = ['optimal lag order'],
).T
```

## Johansen cointegration test

```python
coint_res = find_optimal_coint_rank(
    mvf,
    det_order=1,
    k_ar_diff=10,
    train_only=True,
)
print(coint_res)
coint_res.rank
```

We found a cointegration rank of 0.

## Run VECM
- Now, we can specify a grid that will try more lags, deterministic terms, seasonal fluctuations, and cointegration ranks of 0 and 1

```python
vecm_grid = dict(
    lags = [0], # required to set this to 0 for the vecm model in scalecast
    freq = ['B'], # only necessary to suppress a warning
    k_ar_diff = range(1,66),
    coint_rank = [0,1],
    deterministic = ["n","co","lo","li","cili","colo"],
    seasons = [0,5,30,65,260],
)

mvf.set_estimator('vecm')
mvf.ingest_grid(vecm_grid)
mvf.limit_grid_size(100,random_seed=20)
mvf.cross_validate(k=3)
mvf.auto_forecast()

print(mvf.best_params)

results = mvf.export('model_summaries')
results[[
    'ModelNickname',
    'Series',
    'TestSetMAPE',
    'TestSetMAE',
]]
```

## View VECM Results

```python
results['TestSetMAPE'].mean()
```

```python
mvf.export_validation_grid('vecm').sample(15)
```

```python
mvf.plot_test_set(
    series='AMZN',
    models='vecm',
    include_train=130,
    figsize=(16,8)
)
plt.show()
```

## Re-weight Evaluation Metrics and Rerun VECM

```python
weights = results['TestSetMAPE'] / results['TestSetMAPE'].sum()
weights
```

```python
mvf.add_optimizer_func(
    lambda x: (
        x[0]*weights[0] + 
        x[1]*weights[1] + 
        x[2]*weights[2] + 
        x[3]*weights[3]
    ),
    called = 'weighted',
)
mvf.set_optimize_on('weighted')
mvf.ingest_grid(vecm_grid)
mvf.limit_grid_size(100,random_seed=20)
mvf.cross_validate(k=3)
mvf.auto_forecast(call_me='vecm_weighted')

print(mvf.best_params)

results = mvf.export('model_summaries')
results[[
    'ModelNickname',
    'Series',
    'TestSetMAPE',
    'TestSetMAE',
]]
```

```python
results.loc[results['ModelNickname'] == 'vecm_weighted','TestSetMAPE'].mean()
```

## Try Other MV Models

```python
for stock,f in zip(FANG,fs):
    f.set_validation_length(65)
    f.auto_Xvar_select(
        estimator='rf',
        max_depth=2,
        max_ar=0
    )
    print(f'the ideal representation for {stock} is {f.get_regressor_names()}')
```

```python
mvf2 = MVForecaster(*fs,merge_Xvars='union',names=FANG)
mvf2.set_test_length(65)
mvf2.set_validation_metric('mae')
mvf2
```

```python
GridGenerator.get_mv_grids()
# open MVGrids.py and manually change all lags arguments to range(1,66)
```

```python
models = (
    'mlr',
    'elasticnet',
    'gbt',
    'xgboost',
    'lightgbm',
    'knn',
)

mvf2.tune_test_forecast(
    models,
    limit_grid_size=50,
    cross_validate=True,
    k=3,
)

results = mvf2.export('model_summaries')
results[[
    'ModelNickname',
    'Series',
    'LevelTestSetMAPE',
    'LevelTestSetMAE',
    'Lags',
    'HyperParams',
]]
```
