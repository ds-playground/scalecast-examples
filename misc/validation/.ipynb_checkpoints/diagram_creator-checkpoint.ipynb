```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set(rc={"figure.figsize": (12, 8)})
```

```python
obs = 100
test_length = 10
colors = ['#CCB6F9','#92E985']

x = pd.DataFrame(
    {
        'Fold':[0,1,2,3,4][::-1],
        'Train Data':[75,60,45,30,15][::-1],
        'Validation Data':[15]*5,
    }
)

y = pd.DataFrame(
    {
        'Fold':[0,1,2,3,4][::-1],
        'Dropped':[60,45,30,15,0][::-1],
        'Train Data':[15]*5,
        'Validation Data':[15]*5,
    }
)
```

```python
x
```

```python
x.set_index('Fold').plot.barh(stacked=True,color=colors)
plt.savefig('cv.png')
plt.show()
```

```python
y.set_index('Fold').plot.barh(stacked=True,color=['white'] + colors, legend=False)


backwards = list(range(y.shape[0]))
for i, v in y.iterrows():
    plt.text(v['Train Data'] + v['Dropped'] - 15 + 4,backwards[i] - .1,'Train',color='grey',size=14)
    plt.text(v['Train Data'] + v['Dropped'] + 4,backwards[i] - .1,'Val',color='grey',size=14)

plt.title('Rolling Time Series Cross-Validation',size=20)
plt.ylabel('Fold',size=14)
plt.savefig('rolling_cv.png')
plt.show()
```
