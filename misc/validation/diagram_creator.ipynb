```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
obs = 100
test_length = 10
colors = ['#CCB6F9','#92E985']

x = pd.DataFrame(
    {
        'Fold':[0,1,2,3,4][::-1],
        'Train Data':[75,60,45,30,15][::-1],
        'Validation Data':[15]*5,
    }
)

y = pd.DataFrame(
    {
        'Fold':[0,1,2,3,4][::-1],
        'Dropped':[60,45,30,15,0][::-1],
        'Train Data':[15]*5,
        'Validation Data':[15]*5,
    }
)

z = pd.DataFrame(
    {
        'Fold':[0,1,2,3,4][::-1],
        'Train Data':[72,54,36,18,0][::-1],
        'Validation Data':[18]*5,
        'Dropped':[72,54,36,18,0],
    }
)
```

```python
_, ax = plt.subplots()
x.set_index('Fold').plot.barh(stacked=True,color=colors, legend=False, figsize=(12,8),ax=ax)

backwards = list(range(x.shape[0]))
for i, v in x.iterrows():
    plt.text(4,backwards[i] - .1,'Train',color='#615F66',size=14)
    plt.text(v['Train Data']+ 4,backwards[i] - .1,'Val',color='#615F66',size=14)

ax.axvline(75,color = 'white',ymax=1)
ax.axvline(60,color = 'white',ymax=.8)
ax.axvline(45,color = 'white',ymax=.6)
ax.axvline(30,color = 'white',ymax=.4)
ax.axvline(15,color = 'white',ymax=.2)
    
plt.title('Time Series Cross-Validation',size=20)
plt.ylabel('Fold',size=16)
plt.xlabel('Time Observation Index',size=16)
plt.savefig('cv.png')
plt.show()
```

```python
_, ax = plt.subplots()
y.set_index('Fold').plot.barh(stacked=True,color=['white'] + colors, legend=False, figsize=(12,8),ax=ax)


backwards = list(range(y.shape[0]))
for i, v in y.iterrows():
    plt.text(v['Train Data'] + v['Dropped'] - 15 + 4,backwards[i] - .1,'Train',color='#615F66',size=14)
    plt.text(v['Train Data'] + v['Dropped'] + 4,backwards[i] - .1,'Val',color='#615F66',size=14)
    ax.axvline(15*(i+1),color = 'white')

plt.title('Rolling Time Series Cross-Validation',size=20)
plt.ylabel('Fold',size=16)
plt.xlabel('Time Observation Index',size=16)
plt.savefig('rolling_cv.png')
plt.show()
```

```python
_, ax = plt.subplots()
z.set_index('Fold').plot.barh(stacked=True,color=colors + [colors[0]], legend=False, figsize=(12,8), ax = ax)


backwards = list(range(y.shape[0]))
for i, v in z.iterrows():
    plt.text(v['Train Data'] + 1,backwards[i] - .1,'Validation Slice',color='#615F66',size=14)
    ax.axvline(v['Dropped'],color = 'white')

plt.title('Normal Cross-Validation',size=20)
plt.ylabel('Fold',size=16)
plt.xlabel('Data Split into 5 equal parts',size=14)
plt.xticks([])
plt.savefig('nts_cv.png')
plt.show()
```
