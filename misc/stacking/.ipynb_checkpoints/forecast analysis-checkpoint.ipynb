```python
import pandas as pd
from scalecast.Forecaster import Forecaster
from scalecast.auxmodels import auto_arima
from keras import backend
import matplotlib.pyplot as plt
```

```python
def mean_activation(x):
    return backend.mean(x)
```

```python
def read_data(idx = 'H1'):
    info = pd.read_csv(
        'M4-info.csv',
        index_col=0,
        parse_dates=['StartingDate'],
        dayfirst=True,
    )
    train = pd.read_csv(
        f'Hourly-train.csv',
        index_col=0,
    ).loc[idx]
    test = pd.read_csv(
        f'Hourly-test.csv',
        index_col=0,
    ).loc[idx]
    y = train.to_list() + test.to_list()
    sd = info.loc[idx,'StartingDate']
    fcst_horizon = info.loc[idx,'Horizon']
    cd = pd.date_range(
        start = sd,
        freq = 'H',
        periods = len(y),
    )
    return Forecaster(
        y = y,
        current_dates = cd,
        future_dates = fcst_horizon,
        test_length = fcst_horizon,
    )
```

```python
f = read_data()
f
```

## EDA

```python
f.plot()
plt.show()
```

### ACF/PACF at Series Level

```python
figs, axs = plt.subplots(2, 1,figsize=(9,9))
f.plot_acf(ax=axs[0],title='ACF',lags=48)
f.plot_pacf(ax=axs[1],title='PACF',lags=48)
plt.show()
```

### Augmented Dickey-Fuller Test

```python
critical_pval = 0.05
print('Augmented Dickey-Fuller results:')
stat, pval, _, _, _, _ = f.adf_test(full_res=True)
print('the test-stat value is: {:.2f}'.format(stat))
print('the p-value is {:.4f}'.format(pval))
print('the series is {}'.format('stationary' if pval < critical_pval else 'not stationary'))
print('-'*100)
```

### ACF/PACF at Series First Difference

```python
figs, axs = plt.subplots(2, 1,figsize=(9,9))
f.plot_acf(ax=axs[0],title='ACF - First Diff',lags=48,diffy=1)
f.plot_pacf(ax=axs[1],title='PACF - First Diff',lags=48,diffy=1)
plt.show()
```

### Seasonal Decomp

```python
plt.rc("figure",figsize=(10,6))
f.seasonal_decompose().plot()
plt.show()
```

## ARIMA

### Manual ARIMA: (5,1,4) x (1,1,1,24)

```python
f.set_estimator('arima')
f.manual_forecast(
    order = (5,1,4),
    seasonal_order = (1,1,1,24),
    call_me = 'manual_arima',
)
```

### Auto ARIMA

```python
auto_arima(f,m=24,call_me='auto_arima')
```

```python
f.history['auto_arima']['HyperParams']
```

## RNN

### Mean Activation

```python
f.set_estimator('rnn')
f.manual_forecast(
    lags = 48,
    layers_struct=[
        ('LSTM',{'units':100,'activation':mean_activation}),
        ('LSTM',{'units':100,'activation':mean_activation}),
        ('LSTM',{'units':100,'activation':mean_activation}),
    ],
    optimizer = 'Adam',
    epochs = 15,
    plot_loss = True,
    validation_split=0.2,
    call_me = 'rnn_custom_activation',
)
```

### Tanh Activation

```python
f.manual_forecast(
    lags = 48,
    layers_struct=[
        ('LSTM',{'units':100,'activation':'tanh'}),
        ('LSTM',{'units':100,'activation':'tanh'}),
        ('LSTM',{'units':100,'activation':'tanh'}),
    ],
    optimizer = 'Adam',
    epochs = 15,
    plot_loss = True,
    validation_split=0.2,
    call_me = 'rnn_tanh_activation',
)
```

### Relu Activation

```python
f.manual_forecast(
    lags = 48,
    layers_struct=[
        ('LSTM',{'units':100,'activation':'relu'}),
        ('LSTM',{'units':100,'activation':'relu'}),
        ('LSTM',{'units':100,'activation':'relu'}),
    ],
    optimizer = 'Adam',
    epochs = 15,
    plot_loss = True,
    validation_split=0.2,
    call_me = 'rnn_relu_activation',
)
```
